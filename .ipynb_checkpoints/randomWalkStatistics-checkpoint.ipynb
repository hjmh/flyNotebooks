```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')

%matplotlib inline
```

### Load tracking data (np array)

```python
baseDir = '/Volumes/jayaramanlab/Hannah/FreeWalkingArena/Data/emptyArena/fWTB/f01/'
flyID = 'fWTB01'
arena = 'empty'
```

```python
oneFly = np.load(baseDir + flyID +'_' + arena + '_raw.npy')
```

###### Conversion to metric units, adjustment of time stamps

```python
pixel2mm = 0.471; # 480px, 226mm arena
videoSF = 20 #20Hz video sampling
```

```python
time = oneFly[:,0]
x = oneFly[:,1]
y = oneFly[:,2]
alpha = oneFly[:,3]
```

### Basic processing of tracking data

###### Trace

```python
arrowScale = 5

fig = plt.Figure()
fig.set_canvas(plt.gcf().canvas)
ax = plt.axes()

ax.plot(x,y)
for k in range(1,len(x)):
    ax.arrow(x[k],y[k], np.cos(alpha[k])*arrowScale, np.sin(alpha[k])*arrowScale,
             head_width=0.05, head_length=0.1, fc='grey', ec='grey')

plt.xlabel('x coord [mm]')
plt.ylabel('y coord [mm]')
plt.title('10 min trace of ' + flyID + ' in ' + arena + ' arena')
plt.axis([-10, 490*pixel2mm, -10, 490*pixel2mm])
plt.axes().set_aspect('equal')

plt.show()
```

###### Velocity distributions

```python
transV = np.hstack((0,np.sqrt(np.square(np.diff(x)) + np.square(np.diff(y)) ) * videoSF))
rotV = np.hstack((0, np.diff(alpha) * videoSF))
```

```python
fig = plt.figure(figsize=(11, 6))
fig.set_canvas(plt.gcf().canvas)

plt.subplot(2,3,(1,2))
plt.plot(transV)
plt.ylabel('transl. velocity\n [mm/s]')
plt.ylim((0,50))

plt.subplot(2,3,(4,5))
plt.plot(rotV)
plt.ylabel('rot. velocity\n [rad/s]')
plt.ylim((-20,20))
plt.xlabel('time [s]')

plt.subplot(2,3,3)
transVHist = plt.hist(transV,bins = 50, range = (0, 50), orientation = 'horizontal')
plt.subplot(2,3,6)
rotVHist = plt.hist(rotV,bins = 50, range = (-20, 20), orientation = 'horizontal')
plt.xlabel('count')
```

###### Mean Squared Displacement (MSD)

$$MSD = \langle r^2(\tau) \rangle  = \langle [ r(t + \tau) - r(t)] ^2 \rangle $$

```python
flyPos = np.vstack((x,y))
start = np.asarray(flyPos[:,0])
displacement =  flyPos.T - np.tile(start,[len(x),1]) #vector from fly to arena center
#displacement = np.square(displacement[:,0]) + np.square(displacement[:,1])
```

```python
maxTau = 120
MSD = np.zeros(maxTau*videoSF)
maxSample = 7*60*videoSF
for tau in range(0,maxTau*videoSF): #max is one minute
    sampleDisplacement = np.square( displacement[tau:(maxSample+tau)] - displacement[0:maxSample] )
    MSD[tau] = np.mean(sampleDisplacement)
    
tauSec = np.linspace(0,maxTau,maxTau*videoSF)
```

```python
plt.loglog(tauSec,MSD)
plt.xlabel('tau')
plt.ylabel('MSD')
```

###### Average path length

Sharp turns segment trajectory into path segments

```python
my_cmap = plt.get_cmap('jet')

fig = plt.figure(figsize=(7,7))
fig.set_canvas(plt.gcf().canvas)
ax=plt.axes()

ax.scatter(x,y,c=transV,cmap=my_cmap,vmin=0,vmax=40)
plt.xlabel('x coord [mm]')
plt.ylabel('y coord [mm]')
plt.title('10 min trace of ' + flyID + ' in ' + arena + ' arena')
plt.axis([-10, 490*pixel2mm, -10, 490*pixel2mm])
plt.axes().set_aspect('equal')
plt.colorbar

fig = plt.figure(figsize=(7,7))
fig.set_canvas(plt.gcf().canvas)
ax=plt.axes()

ax.scatter(x,y,c=abs(rotV),cmap=my_cmap,vmin=0,vmax=15)
plt.xlabel('x coord [mm]')
plt.ylabel('y coord [mm]')
plt.title('10 min trace of ' + flyID + ' in ' + arena + ' arena')
plt.axis([-10, 490*pixel2mm, -10, 490*pixel2mm])
plt.axes().set_aspect('equal')
plt.colorbar
```

```python
turns = (transV<20).astype('int') & (abs(rotV)>1).astype('int')
turns = turns.astype('bool')

arenaCenter = np.asarray((240*pixel2mm, 240*pixel2mm))
distCenter =  flyPos.T - np.tile(arenaCenter,[len(x),1]) #vector from fly to arena center
distCenter = np.linalg.norm(np.asarray(distCenter),axis=1)

plt.figure(figsize=(10,10))
plt.plot(x,y,c='grey')
plt.scatter(x[turns],y[turns])
plt.scatter(x[distCenter> (220*pixel2mm)],y[distCenter> (220*pixel2mm)],c='red')
plt.axis([-10, 490*pixel2mm, -10, 490*pixel2mm])
plt.axes().set_aspect('equal')
```

```python
pathID = np.zeros(len(x))
inpath = True
currID = 0
for t in range(0,len(x)):
    if turns[t]:
        pathID[t] = float('nan')
        if inpath:
            inpath = False
            currID = currID + 1
    else:
        if not inpath:
            inpath = True
        pathID[t] = currID
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y,c=pathID,cmap = plt.get_cmap('Set3'))
```
