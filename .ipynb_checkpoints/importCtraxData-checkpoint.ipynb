### Import ctrax tracking data

```python
from scipy.io import loadmat
import csv
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
```

```python
baseDir = '/Volumes/jayaramanlab/Hannah/FreeWalkingArena/Data/emptyArena/fWTB/f01/'
flyID = 'fWTB01'
arena = 'empty'
infile = ['f01WTB_empty_7d_2015-03-12-165328-000'+ str(x) + '.mat' for x in range(0,4)]
```

```python
keyList = ['timestamps','x_pos','y_pos','angle'] #data columns to be extracted from ctrax file
datCombi = np.empty([0,4])
```

```python
for f in range(0,4):
    indat = loadmat(baseDir + infile[f])
    mask = np.where(indat['ntargets'] == 1)[0]
    dat = [indat[k] for k in keyList]
    for ind,val in enumerate(dat):
        if len(val) > len(mask):
            dat[ind] = val[mask]

    dat = np.squeeze(np.asarray(dat))
    datCombi = np.vstack((datCombi,dat.T))
```

###### Conversion to metric units, adjustment of time stamps

```python
pixel2mm = 0.471; # pixel2mm (480px, 226mm arena) and 20Hz video sampling
videoSF = 20
```

```python
datCombi[:,1] = datCombi[:,1]*pixel2mm
datCombi[:,2] = np.abs(datCombi[:,2]-480)*pixel2mm #flip y axis (artifact from image processing)

datCombi[:,3] = np.mod((2*np.pi - datCombi[:,3]),2*np.pi) #transform orientation angle to account for flipped y-axis
```

```python
# TODO: Adjust time stamps
plt.plot(datCombi[:,0])
```

```python
with open(baseDir + flyID +'_' + arena + '_raw.csv', 'wb') as csvfile:
    matwriter = csv.writer(csvfile) # delimiter=' ', quotechar='\n', quoting=csv.QUOTE_MINIMAL)
    matwriter.writerows(datCombi)
```

### Basic processing of tracking data

###### Trace

```python
arrowScale = 5

fig = plt.Figure()
fig.set_canvas(plt.gcf().canvas)
ax = plt.axes()

ax.plot(datCombi[:,1], datCombi[:,2])
for k in range(1,len(datCombi[:,1])):
    ax.arrow(datCombi[k,1], datCombi[k,2], np.cos(datCombi[k,3])*arrowScale, np.sin(datCombi[k,3])*arrowScale,
             head_width=0.05, head_length=0.1, fc='grey', ec='grey')

plt.xlabel('x coord [mm]')
plt.ylabel('y coord [mm]')
plt.title('10 min trace of ' + flyID + ' in ' + arena + ' arena')
plt.axis([-10, 490*pixel2mm, -10, 490*pixel2mm])
plt.axes().set_aspect('equal')

plt.show()
```

```python
fig.savefig(baseDir + flyID + '_' + arena + '_trace.pdf', format = 'pdf')
```

###### Velocity distributions

```python
transV = np.sqrt(np.square(np.diff(datCombi[:,1])) + np.square(np.diff(datCombi[:,2])) ) * videoSF
rotV = np.diff(datCombi[:,3]) * videoSF
```

```python
fig = plt.figure()
fig.set_canvas(plt.gcf().canvas)
plt.subplot(2,1,1)
plt.plot(transV)
plt.ylabel('transl. velocity\n [mm/s]')
plt.ylim((0,70))

plt.subplot(2,1,2)
plt.plot(rotV)
plt.ylabel('rot. velocity\n [rad/s]')
plt.ylim((-50,50))
plt.xlabel('time [s]')
```

```python
fig = plt.figure()
plt.subplot(121)
transVHist = plt.hist(transV,bins = 50, range = (0, 70))
plt.ylabel('count')
plt.xlabel('transl. velocity\n [mm/s]')
plt.subplot(122)
rotVHist = plt.hist(rotV,bins = 50, range = (-20, 20))
plt.xlabel('rot. velocity\n [rad/s]')
```

###### Headedness

```python
arenaCenter = np.asarray((240*pixel2mm, 240*pixel2mm))
flyPos = np.vstack((datCombi[:,1],datCombi[:,2]))
dCenter =  flyPos.T - np.tile(arenaCenter,[len(datCombi[:,1]),1]) #vector from fly to arena center

#angular position from arena center: 0 to pi on upper half...
angPos =  np.arccos(dCenter[:,0] / np.sqrt(np.square(dCenter[:,0]) + np.square(dCenter[:,1])) )
angPos[dCenter[:,1] < 0] = 2*np.pi - angPos[dCenter[:,1] < 0];
```

```python
orientationScore = abs(angPos - datCombi[:,3])
```

```python
fig = plt.figure()
plt.scatter(angPos,datCombi[:,3])
plt.xlabel('angular position of fly'); plt.ylabel('orientation of fly'); 
```

```python
heading = np.arctan(np.diff(datCombi[:,2])/np.diff(datCombi[:,1]))
heading = np.hstack((heading[1],heading))
circlingScore = abs(angPos - heading+np.pi)
```

```python
plt.hist(circligScore,bins = 50);
```
