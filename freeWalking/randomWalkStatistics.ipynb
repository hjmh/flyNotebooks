###### Velocity distributions

```python
transV = np.hstack((0,np.sqrt(np.square(np.diff(x)) + np.square(np.diff(y)) ) * videoSF))
rotV = np.hstack((0, np.diff(alpha) * videoSF))
```

```python
fig = plt.figure(figsize=(11, 6))
fig.set_canvas(plt.gcf().canvas)

plt.subplot(2,3,(1,2))
plt.plot(transV)
plt.ylabel('transl. velocity\n [mm/s]')
plt.ylim((0,50))

plt.subplot(2,3,(4,5))
plt.plot(rotV)
plt.ylabel('rot. velocity\n [rad/s]')
plt.ylim((-20,20))
plt.xlabel('time [s]')

plt.subplot(2,3,3)
transVHist = plt.hist(transV,bins = 50, range = (0, 50), orientation = 'horizontal')
plt.subplot(2,3,6)
rotVHist = plt.hist(rotV,bins = 50, range = (-20, 20), orientation = 'horizontal')
plt.xlabel('count')
```

### Mean Squared Displacement (MSD)

$$MSD = \langle r^2(\tau) \rangle  = \langle [ r(t + \tau) - r(t)] ^2 \rangle $$

```python
flyPos = np.vstack((x,y))
start = np.asarray(flyPos[:,0])
displacement =  flyPos.T - np.tile(start,[len(x),1]) #vector from fly to arena center
#displacement = np.square(displacement[:,0]) + np.square(displacement[:,1])
```

```python
maxTau = 120
MSD = np.zeros(maxTau*videoSF)
maxSample = 7*60*videoSF
for tau in range(0,maxTau*videoSF): #max is one minute
    sampleDisplacement = np.square( displacement[tau:(maxSample+tau)] - displacement[0:maxSample] )
    MSD[tau] = np.mean(sampleDisplacement)
    
tauSec = np.linspace(0,maxTau,maxTau*videoSF)
```

```python
plt.figure(figsize=(5,5))
plt.loglog(tauSec,MSD)
plt.xlabel('tau')
plt.ylabel('MSD')
plt.axes().set_aspect('equal')
```

### Distribution of path length

Sharp turns segment trajectory into path segments

```python
# Combine turn definition, plot and widgets
```

```python
turns = (transV<20).astype('int') & (abs(rotV)>3).astype('int')
turns = turns.astype('bool')

arenaCenter = np.asarray((240*pixel2mm, 240*pixel2mm))
distCenter =  flyPos.T - np.tile(arenaCenter,[len(x),1]) #vector from fly to arena center
distCenter = np.linalg.norm(np.asarray(distCenter),axis=1)
```

```python
pathID = np.zeros(len(x))
inpath = True
currID = 0
for t in range(0,len(x)):
    if turns[t]:
        pathID[t] = float('nan')
        if inpath:
            inpath = False
            currID = currID + 1
    else:
        if not inpath:
            inpath = True
        pathID[t] = currID
```

```python
pathLength = np.zeros(max(pathID))
for path in range(0,max(pathID).astype('int')):
    currPath = np.argwhere(pathID==path)
    pathLength[path] = sum(np.sqrt(np.square(x[currPath[0:-2]]-x[currPath[1:-1]])
                                   + np.square(y[currPath[0:-2]]-y[currPath[1:-1]]))) 
```

```python
fig = plt.figure(figsize=(10,15))
sns.set_style('whitegrid')
sns.set_palette('RdBu')

fig.add_subplot(3,1,(1,2), adjustable = 'box', aspect = 1,
                    xlim = (0,480*pixel2mm), ylim = (0,480*pixel2mm))
#plt.scatter(x[distCenter> (220*pixel2mm)],y[distCenter> (220*pixel2mm)],c='red', alpha = 0.3)
plt.scatter(x,y,c=pathID, cmap='flag',alpha = 0.5)
plt.scatter(x[turns],y[turns],color = 'black')
plt.xlabel('x coord [mm]'); plt.ylabel('y coord [mm]')

fig.add_subplot(3,1,3,adjustable = 'box', aspect = 0.4)
pathLength_filt = pathLength[pathLength>2]
plt.hist(pathLength_filt,bins=50);
plt.xlabel('path length [mm]'); plt.ylabel('count')
```

###### Translational velocity vs. path length

```python
pathVelo = np.zeros(max(pathID))
for path in range(0,max(pathID).astype('int')):
    currPath = np.argwhere(pathID==path)
    pathVelo[path] = np.mean(transV[currPath])
```

```python
fig = plt.figure(figsize = (10,5))
plt.scatter(pathVelo, pathLength,alpha = 0.3)
plt.xlabel('mean transl. velocity [mm/s]')
plt.ylabel('path length [mm]')
```
