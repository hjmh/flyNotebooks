###### Distribution of path length

Sharp turns segment trajectory into path segments

```python
def computePathLength(turns,x,y,nPts):
    pathID = np.zeros(nPts)
    inpath = True
    currID = 0
    for t in range(0,nPts):
        if turns[t]:
            pathID[t] = float('nan')
            if inpath:
                inpath = False
                currID = currID + 1
        else:
            if not inpath:
                inpath = True
            pathID[t] = currID
    
    pathLength = np.zeros(max(pathID))
    for path in range(0,max(pathID).astype('int')):
        currPath = np.argwhere(pathID==path)
        pathLength[path] = sum(np.hypot(x[currPath[0:-2]]-x[currPath[1:-1]],y[currPath[0:-2]]-y[currPath[1:-1]]))
        
    return pathID,pathLength
```

```python
pathID = []
pathLength = []
turns = []

for flyInd, flyID in enumerate(flyIDlist):
    currFly = allFlies[:,0] == flyInd
    
    x = allx[currFly]
    y = ally[currFly]
    tV = alltV[currFly]
    rV = allrV[currFly] * 180.0/np.pi
    
    flyPos = np.vstack((allx[currFly],ally[currFly]))
    
    currTurns = (tV<20).astype('int') & (abs(rV)>10).astype('int')
    currTurns = currTurns.astype('bool')

    #arenaCenter = np.asarray((240*pixel2mm, 240*pixel2mm))
    #distCenter =  flyPos.T - np.tile(arenaCenter,[len(tV),1]) #vector from fly to arena center
    #distCenter = np.linalg.norm(np.asarray(distCenter),axis=1)
    
    pathsLengths = computePathLength(currTurns,x,y,len(x))
    pathID.append(pathsLengths[0])
    pathLength.append(pathsLengths[1])
    turns.append(currTurns)
```

### Functions

```python
def loadCtraxData(flyID,arena):
    oneFly = np.load(baseDir + flyID[0] + flyID[-2:] + '/' + flyID +'_' + arena + '_raw.npy')
    time = oneFly[:,0]
    x = oneFly[:,1]
    y = oneFly[:,2]
    alpha = oneFly[:,3]
    
    return (time,x,y,alpha)
```

```python
def computeMSD(x,y,maxTau,videoSF):
    flyPos = np.vstack((x,y))
    start = np.asarray(flyPos[:,0])
    displacement =  flyPos.T - np.tile(start,[len(x),1]) #vector from fly to arena center

    MSD = np.zeros(maxTau*videoSF)
    maxSample = 7*60*videoSF
    for tau in range(0,maxTau*videoSF): #max is one minute
        sampleDisplacement = np.square( displacement[tau:(maxSample+tau)] - displacement[0:maxSample] )
        MSD[tau] = np.mean(sampleDisplacement)
    
    return MSD
```
